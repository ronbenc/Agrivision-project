# Helper Functions

```python
#creates a list of files based on one-hot encoded file


def sort_folder_into_classes(encoded_df, column_name, list_name):
    
    
    list_name = encoded_df.loc[encoded_df[column_name] == 1] #select columns you want to move
    list_name['index'] = list_name['index'].astype(str) + '.jpg' #make sure the .jpg lines up as original files are png but directory we are matching is jpeg

    identifier_list = [] #create am empty list for the identiies

    identifier_list = list_name['index'] #create a list of relevant image names
    identifier_list.append(identifier_list) #append to empty list
    unique_identifiers = set(identifier_list) #transform to a set 

    return unique_identifiers 
```

```python
# move folders 


def move_unique_id_to_folder(mypath, unique_identifier_set):

#path to image folder is specified in argument
# onlyfiles stores it in a list

    onlyfiles = [f for f in os.listdir(mypath) if os.path.isfile(os.path.join(mypath, f))]

    #ls1 is a list of unique identifiers generated by the previous function
    ls1 = unique_identifiers

    #list of image names
    ls1files = [i for i in onlyfiles for j in ls1 if j in i]

    subfolder1 = os.path.join(mypath, 'new_folder_name'')

    # check if they already exits to prevent error
    if not os.path.exists(subfolder1):
    os.makedirs(subfolder1)

    # move files to their respective sub folders
    for i in ls1files:
    source = os.path.join(mypath, i)
    destination = os.path.join(subfolder1, i)
    os.rename(source, destination)
    
    return print('Check Destination folder to see your newly sorted files!') 
```
